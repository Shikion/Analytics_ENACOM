```python
import pandas as pd
```

```python
hist_vel = pd.read_csv("datasets/historico_velocidad_internet.csv")
hist_vel.head()
```

```python
hist_vel = hist_vel.drop("Trimestre.1", axis=1)
hist_vel.head()
```

```python
hist_vel['Mbps (Media de bajada)'] = hist_vel['Mbps (Media de bajada)'].str.replace(',', '.')
hist_vel["Mbps (Media de bajada)"] = hist_vel["Mbps (Media de bajada)"].astype(float)
hist_vel.head()
```

```python
BAF = pd.read_csv("datasets/Internet_BAF.csv")
BAF.head()
```

```python
BAF['Banda ancha fija'] = BAF['Banda ancha fija'].str.replace('.', '')
BAF['Total'] = BAF['Total'].str.replace('.', '')
```

```python
BAF['Banda ancha fija'] = BAF['Banda ancha fija'].astype(int)
BAF['Total'] = BAF['Total'].astype(int)
```

```python
BAF["Dial up"] = BAF["Dial up"].astype(str)
BAF["Dial up"] = BAF["Dial up"].str.replace('.', '')
BAF["Dial up"] = BAF["Dial up"].astype(int)
```

```python
BAF = BAF.drop("Periodo", axis=1)
BAF.head()
```

```python
penetracion = pd.read_csv("datasets/Internet_Penetracion.csv")
penetracion = penetracion.iloc[::-1]
penetracion.head()
```

```python
penetracion["Accesos por cada 100 hogares"] = penetracion["Accesos por cada 100 hogares"].str.replace(',', '.')
```

```python
penetracion["Accesos por cada 100 hogares"] = penetracion["Accesos por cada 100 hogares"].astype(float)
penetracion.head()
```

```python
tec = pd.read_csv("datasets/Internet_Accesos-por-tecnologia.csv")
tec = tec.drop("Periodo", axis=1)
```

```python
tec["ADSL"] = tec["ADSL"].str.replace('.', '')
tec["Cablemodem"] = tec["Cablemodem"].str.replace('.', '')
tec["Fibra óptica"] = tec["Fibra óptica"].str.replace('.', '')
tec["Total"] = tec["Total"].str.replace('.', '')
```

```python
tec["ADSL"] = tec["ADSL"].astype(int)
tec["Cablemodem"] = tec["Cablemodem"].astype(int)
tec["Fibra óptica"] = tec["Fibra óptica"].astype(int)
tec["Total"] = tec["Total"].astype(int)
tec.head()
```

```python
tec["Wireless"] = tec["Wireless"].astype(str)
tec["Wireless"] = tec["Wireless"].str.replace('.', '')
tec["Wireless"] = tec["Wireless"].astype(int)
```

```python
tec["Otros"] = tec["Otros"].astype(str)
tec["Otros"] = tec["Otros"].str.replace('.', '')
tec["Otros"] = tec["Otros"].astype(int)
tec.head()
```

```python
prov = pd.read_csv("datasets/Internet_Accesos-por-velocidad.csv")
prov.head()
```

```python
prov["HASTA 512 kbps"] = prov["HASTA 512 kbps"].astype(str)
prov["HASTA 512 kbps"] = prov["HASTA 512 kbps"].str.replace('.', '')
prov["HASTA 512 kbps"] = prov["HASTA 512 kbps"].astype(int)

prov["+ 512 Kbps - 1 Mbps"] = prov["+ 512 Kbps - 1 Mbps"].astype(str)
prov["+ 512 Kbps - 1 Mbps"] = prov["+ 512 Kbps - 1 Mbps"].str.replace('.', '')
prov["+ 512 Kbps - 1 Mbps"] = prov["+ 512 Kbps - 1 Mbps"].astype(int)

prov["+ 1 Mbps - 6 Mbps"] = prov["+ 1 Mbps - 6 Mbps"].astype(str)
prov["+ 1 Mbps - 6 Mbps"] = prov["+ 1 Mbps - 6 Mbps"].str.replace('.', '')
prov["+ 1 Mbps - 6 Mbps"] = prov["+ 1 Mbps - 6 Mbps"].astype(int)
prov.head()
```

```python
prov["+ 6 Mbps - 10 Mbps"] = prov["+ 6 Mbps - 10 Mbps"].astype(str)
prov["+ 6 Mbps - 10 Mbps"] = prov["+ 6 Mbps - 10 Mbps"].str.replace('.', '')
prov["+ 6 Mbps - 10 Mbps"] = prov["+ 6 Mbps - 10 Mbps"].astype(int)

prov["+ 10 Mbps - 20 Mbps"] = prov["+ 10 Mbps - 20 Mbps"].astype(str)
prov["+ 10 Mbps - 20 Mbps"] = prov["+ 10 Mbps - 20 Mbps"].str.replace('.', '')
prov["+ 10 Mbps - 20 Mbps"] = prov["+ 10 Mbps - 20 Mbps"].astype(int)

prov["+ 20 Mbps - 30 Mbps"] = prov["+ 20 Mbps - 30 Mbps"].astype(str)
prov["+ 20 Mbps - 30 Mbps"] = prov["+ 20 Mbps - 30 Mbps"].str.replace('.', '')
prov["+ 20 Mbps - 30 Mbps"] = prov["+ 20 Mbps - 30 Mbps"].astype(int)
prov.head()
```

```python
prov["+ 30 Mbps"] = prov["+ 30 Mbps"].astype(str)
prov["+ 30 Mbps"] = prov["+ 30 Mbps"].str.replace('.', '')
prov["+ 30 Mbps"] = prov["+ 30 Mbps"].astype(int)

prov["Total"] = prov["Total"].astype(str)
prov["Total"] = prov["Total"].str.replace('.', '')
prov["Total"] = prov["Total"].astype(int)
prov.head()
```

```python
hist_vel['trimestre_año'] = hist_vel['Año'].astype(str) + ' T' + hist_vel['Trimestre'].astype(str)
BAF['trimestre_año'] = BAF['Año'].astype(str) + ' T' + BAF['Trimestre'].astype(str)
penetracion['trimestre_año'] = penetracion['Año'].astype(str) + ' T' + penetracion['Trimestre'].astype(str)
tec['trimestre_año'] = tec['Año'].astype(str) + ' T' + tec['Trimestre'].astype(str)
prov['trimestre_año'] = prov['Año'].astype(str) + ' T' + prov['Trimestre'].astype(str)
```

```python
hist_vel = hist_vel.drop(["Año", "Trimestre"], axis=1)
BAF = BAF.drop(["Año", "Trimestre"], axis=1)
penetracion = penetracion.drop(["Año", "Trimestre"], axis=1)
tec = tec.drop(["Año", "Trimestre"], axis=1)
prov = prov.drop(["Año", "Trimestre"], axis=1)
```

```python
hist_vel.to_csv("datasets/historico_velocidad_internet.csv", index=False)
BAF.to_csv("datasets/Internet_BAF.csv", index=False)
penetracion.to_csv("datasets/Internet_Penetracion.csv", index=False)
tec.to_csv("datasets/Internet_Accesos-por-tecnologia.csv", index=False)
prov.to_csv("datasets/Internet_Accesos-por-velocidad.csv", index=False)
```
